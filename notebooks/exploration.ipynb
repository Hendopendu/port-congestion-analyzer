```python
import sqlite3
import pandas as pd
import os

# Since you are in /app, the path is just 'data/raw/ais_raw.db'
db_path = "data/raw/ais_raw.db"

if os.path.exists(db_path):
    conn = sqlite3.connect(db_path)
    df = pd.read_sql_query("SELECT * FROM ship_positions", conn)
    print(f"✅ Success! Found {len(df)} rows.")
    display(df.tail()) # This shows the last 5 ships you caught
else:
    print(f"❌ File not found at: {os.path.abspath(db_path)}")
    # Let's see what is inside the data/raw folder to be sure
    if os.path.exists("data/raw"):
        print(f"Contents of data/raw: {os.listdir('data/raw')}")
    else:
        print("The folder 'data/raw' doesn't even exist yet. Did you run the ingestor?")
```
